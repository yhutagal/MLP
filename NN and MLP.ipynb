```python
import numpy as np
import pandas as pd
```

```python
#Fungsi Aktivasi
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1-x)
```

```python
def read_csv(file_path, delimiter=';'):
    with open(file_path, 'r') as file:
      lines = file.readlines()
    data = [line.strip().split(delimiter) for line in lines]
    columns = data[0]
    data = data[1:]
    df = pd.DataFrame (data, columns=columns)
    return df
```

```python
train_data = read_csv("dota2Train.csv")
test_data = read_csv("dota2Test.csv")
```

```python
#Preprocess Data
train_data = train_data.apply(pd.to_numeric)
test_data = test_data.apply(pd.to_numeric)
```

```python
X_train = train_data.drop('label', axis=1).values
y_train = train_data['label'].values.reshape(-1,1)
X_test = test_data.drop('label', axis=1).values
y_test = test_data['label'].values.reshape(-1,1)
```

```python
#Normalisasi Fitur
X_train = (X_train - np.mean(X_train)) / np.std(X_train)
X_test = (X_test - np.mean(X_test)) / np.std(X_test)
```

Class Neural Network=



1.   Inisialisasi bobot, input, hidden, ouput
2.   Forward Phase
3.   Backward Phase
4.   Pembaruan Bobot (update weight)
5.   Training

```python
class NeuralNetwork:
  def __init__(self, ukuran_masukan, jaringan_tersembunyi, ukuran_keluaran):
      self.ukuran_masukan = ukuran_masukan
      self.jaringan_tersembunyi = jaringan_tersembunyi
      self.ukuran_keluaran = ukuran_keluaran
      self.hasil_epoch = []

      #membuat bobot dan bias
      self.bobot_masukan_tersembunyi = np.random.randn(self.ukuran_masukan, jaringan_tersembunyi)
      self.bias_masukan_tersembunyi = np.zeros((1, self.jaringan_tersembunyi))
      self.bobot_keluaran_tersembunyi = np.random.randn(self.jaringan_tersembunyi, self.ukuran_keluaran)
      self.bias_keluaran_tersembunyi = np.zeros((1, self.ukuran_keluaran))

  #Fungsi Forward
  def forward(self, x):
      self.masukan_tersembunyi = np.dot(x, self.bobot_masukan_tersembunyi) + self.bias_masukan_tersembunyi
      self.keluaran_tersembunyi = sigmoid(self.masukan_tersembunyi)
      self.keluaran = sigmoid(np.dot(self.keluaran_tersembunyi, self.bobot_keluaran_tersembunyi))
      return self.keluaran

  #Fungsi Backward
  def backward(self, x, y, keluaran, learning_rate):
      error = y - keluaran
      d_keluaran = error * sigmoid_derivative(keluaran)

      error_tersembunyi = np.dot(d_keluaran, self.bobot_keluaran_tersembunyi.T)
      d_tersembunyi = error_tersembunyi * sigmoid_derivative(self.keluaran_tersembunyi)

      #Pembaruan Bobot dan Bias
      self.bobot_keluaran_tersembunyi += np.dot(self.keluaran_tersembunyi.T, d_keluaran) * learning_rate
      self.bias_keluaran_tersembunyi += np.sum(d_keluaran, axis=0, keepdims=True) * learning_rate
      self.bobot_masukan_tersembunyi += np.dot(x.T, d_tersembunyi) * learning_rate
      self.bias_masukan_tersembunyi += np.sum(d_tersembunyi, axis=0, keepdims = True) * learning_rate

  def latih(self, x, y, epochs, learning_rate):
      for epoch in range(epochs):
          keluaran = self.forward(x)
          self.backward(x,y, keluaran, learning_rate)
          if epoch % 100 == 0:
              loss = np.mean(np.square(y-keluaran))
              akurasi = np.mean(np.round(keluaran) == y)
              print(f"epoch {epoch}, loss: {loss}, akurasi: {akurasi}")
```

```python
ukuran_masukan = x_train.shape[1]
jaringan_tersembunyi = 166
ukuran_keluaran = 1
learning_rate = 0.01
epochs = 1000
```

```python
model = NeuralNetwork(ukuran_masukan, jaringan_tersembunyi, ukuran_keluaran)
model.latih(x_train, y_train, epochs, learning_rate)
```
